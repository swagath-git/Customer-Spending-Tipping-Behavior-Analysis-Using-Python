# Joint Plot
1. a joint plot allowa to study the relationship between 2 numerical variable. The central chart display their correlation, It is usually a sctterplot, a hexbin plot, a 2D density plot

```python
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

# Violin Plot
1. Violin plot helps us to see both the distribution of data interms of krnal sensity estimation and the box plot

```python
sns.violinplot(x='total_bill',y='sex',data=df)
plt.show()
```

# Pair Plot _ MUltiple continous variable
1. a "Pairs Plot" is also known as a scatterplot in which one variable in the same data row is matched with another variable's value like this; Pairs plots are just elaboration on this showing all variables papired with all the other variables

```python
sns.pairplot(df,vars=continous)
plt.show()
```

# Heat Map
1. A Heat map uses colored cells to represent relation between variables

### Heatmap(for correlation)
1. A correlation heatmap uses colored cells to show a 2D correlation matrix (table) between two numeric dimensions
2. it is a very important in feature delection

### correlation (r)=

```python
c_m = df[continous].corr()
c_m
```

```python
sns.heatmap(c_m,annot=True)
plt.show()
```
